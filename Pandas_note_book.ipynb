```python
df = pd.read_csv(r"C:\Users\admin\Downloads\archive (74)\flight.csv")
df.sample(5)
```

```python
df = df[["Flight_Date", "UniqueCarrier", "Origin_Airport", "Origin_City_Name", "Origin_State", "Dep_Delay", "Arrival_Delay", "Cancelled", 
         "Diverted", "Distance", "Carrier_Delay", "WeatherDelay", "NAS_Delay", "Security_Delay", "Late_Aircraft_Delay"]]
```

```python
df
```

```python
df.Origin_City_Name= pd.DataFrame(df.Origin_City_Name.str.split(",").str[0])
df.sample(5)
```

```python
state_map = {'MA': 'Massachusetts', 'NY': 'New York', 'CA': 'California', 'TX': 'Texas', 'IL': 'Illinois', 'NJ': 'New Jersey', 'FL': 'Florida', 'GA': 'Georgia', 'NV': 'Nevada', 'UT': 'Utah', 'CO': 'Colorado', 'NC': 'North Carolina', 'AZ': 'Arizona', 'MI': 'Michigan', 'MN': 'Minnesota'}

df["Origin_State"] = df["Origin_State"].map(state_map)

df.sample(5)
```

```python
df.columns
```

```python
df = df.drop("UniqueCarrier", axis = 1)
```

```python
df.sample()
```

```python
df = df.rename({"Distance": "distance_km"}, axis = 1)

df["distance_km"] = round(df["distance_km"] * 1.609, 1)
```

```python
df.sample(5)
```

```python
rename_map = {
    'Carrier_Delay': 'Airline_Delay_Min',
    'WeatherDelay': 'Weather_Delay_Min',
    'NAS_Delay': 'Airport_Traffic_Delay_Min',
    'Security_Delay': 'Security_Check_Delay_Min',
    'Late_Aircraft_Delay': 'Previous_Flight_Delay_Min', 
    'Dep_Delay': 'Departure_Delay_min', 
    'Arrival_Delay': "Arrival_Delay_min"
}

df = df.rename(columns=rename_map)

df.sample(5)
```

```python
print(df.Cancelled.unique(), df.Diverted.unique())
```

```python
df = df.drop(["Cancelled", "Diverted"], axis = 1)

df.sample(1)
```

```python
df.sample(5)
```

```python
df.columns = df.columns.str.lower()
df.sample()
```

```python
df = df.rename(columns={"origin_airport": "airport", "origin_city_name": "city", "origin_state": "state"})
```

```python
df.sample(5)
```

```python
df.sample(5)
```

```python
df["flight_date"] = pd.to_datetime(df["flight_date"], errors = "raise", dayfirst=True)
```
